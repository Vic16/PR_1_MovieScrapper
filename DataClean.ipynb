```python
import pandas as pd
import json
```

```python
with open('source/data_wiki.json', 'r') as f:
    data = json.load(f)
```

```python
moviesData = []
moviesMetadata = []

for e,i in enumerate(data):
    keys = list(data[e].keys())
    if 'title' in keys:
        moviesData.append(data[e])
        #data.pop(e)
    else:
        moviesMetadata.append(data[e])
```

```python
df = pd.DataFrame(moviesData)
```
